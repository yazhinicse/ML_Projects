### b. Code to compute the Projection Matrix

```python
import numpy as np
A = []
for i in range(-5, 6):
 A.append([i, 1])
A = np.array(A)
projection_matrix_A = A.dot(np.linalg.inv(A.T.dot(A))).dot(A.T)
print(f"The projection matrix of \n {A} is \n\n {projection_matrix_A}")
```
